```python
#First we will use RandomOverSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
#splitting input and output
x=df.drop("churn",axis=1)
y=df["churn"]
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#to convert minority class (1) into majority class (0)
ros= RandomOverSampler(random_state=1)
```

```python
#apply over sampler on training data
x_train_ros, y_train_ros=ros.fit_resample(x_train, y_train)
```

```python
y_train_ros.value_counts()
```

```python
#apply over sampler on testing data
x_test_ros,y_test_ros=ros.fit_resample(x_test,y_test)
```

```python
y_test_ros.value_counts()
```

```python
#import LogisticRagresion
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
#create function
#create a function
def create_model(model):          #model user defined which holds the object of algo
    #train the model
    model.fit(x_train_ros,y_train_ros)
    #test the model
    y_pred=model.predict(x_test_ros)
    #Generate report
    print(classification_report(y_test_ros,y_pred))
    print(confusion_matrix(y_test_ros,y_pred))
    return model

#import libraries
from sklearn.metrics import classification_report,confusion_matrix

#call the function
lr=create_model(lr)
```
